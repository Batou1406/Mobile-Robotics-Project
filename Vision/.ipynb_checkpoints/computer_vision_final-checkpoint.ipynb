```python
import cv2

cap = cv2.VideoCapture(0)

# Check if the webcam is opened correctly
if not cap.isOpened():
    raise IOError("Cannot open webcam")

while True:
    ret, frame = cap.read()
    frame = cv2.resize(frame, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA)
    cv2.imshow('Input', frame)

    c = cv2.waitKey(1)
    if c == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#importing the required libraries 
import numpy as np 
import cv2 
import matplotlib.pyplot as plt 
from matplotlib import colors
%matplotlib inline 

kernel_size=5

Map = cv2.imread('MAP_3.jpg') 
Map_blur = cv2.blur(Map,(kernel_size,kernel_size))

# convert to RGB
Map = cv2.cvtColor(Map, cv2.COLOR_BGR2RGB)
plt.figure()
plt.imshow(Map)
# convert to grayscale
gray = cv2.cvtColor(Map, cv2.COLOR_RGB2GRAY)
 
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)
# show it
plt.figure()
plt.imshow(binary, cmap="gray")
plt.show()
```

```python
def create_empty_plot(max_val):
    """
    Helper function to create a figure of the desired dimensions & grid
    
    :param max_val: dimension of the map along the x and y dimensions
    :return: the fig and ax objects.
    """
    fig, ax = plt.subplots(figsize=(7,7))
    
    major_ticks = np.arange(0, max_val+1, 5)
    minor_ticks = np.arange(0, max_val+1, 1)
    ax.set_xticks(major_ticks)
    ax.set_xticks(minor_ticks, minor=True)
    ax.set_yticks(major_ticks)
    ax.set_yticks(minor_ticks, minor=True)
    ax.grid(which='minor', alpha=0.2)
    ax.grid(which='major', alpha=0.5)
    ax.set_ylim([-1,max_val])
    ax.set_xlim([-1,max_val])
    ax.grid(True)
    
    return fig, ax
```

```python
#Creating the grid
max_val = 50 # Size of the map

fig, ax = create_empty_plot(max_val)

# Creating the occupancy grid
#np.random.seed(0) # To guarantee the same outcome on all computers
#data = np.random.rand(max_val, max_val) * 20 # Create a grid of 50 x 50 random values
#cmap = colors.ListedColormap(['white', 'red']) # Select the colors with which to display obstacles and free cells

# more simple to have obstacle in black and free path in white (pixels equal 0)
binary[binary==0] = 127
binary[binary==255] = 0
binary[binary==127] = 255

occupancy_grid = np.zeros((50,50))

for i in range(50):
    for j in range(50):
        result = np.sum(binary[24*i:24*(i+1),24*j:24*(j+1)])
        if result > 0 : 
            occupancy_grid[i,j] = 1

# Displaying the map
plt.imshow(occupancy_grid, cmap="gray")
plt.show()
```
