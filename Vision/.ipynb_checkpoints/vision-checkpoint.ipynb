```python
#importing the required libraries 
import numpy as np 
import cv2 
import matplotlib.pyplot as plt 
from matplotlib import colors
%matplotlib inline 
image = cv2.imread('map_obstacles.jpg') 
# convert to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)
# show it
plt.imshow(binary, cmap="gray")

plt.show()
```

```python
# more simple to have obstacle in black and free path in white (pixels equal 0)
binary[binary==0] = 127
binary[binary==255] = 0
binary[binary==127] = 255

occupancy_grid = np.zeros((50,50))

for i in range(50):
    for j in range(50):
        result = np.sum(binary[24*i:24*(i+1),24*j:24*(j+1)])
        if result > 0 : 
            occupancy_grid[i,j] = 1
            
plt.imshow(occupancy_grid, cmap="gray")
plt.show()  
```

```python
#tuto4 à rajouter source

import cv2
import numpy as np
from matplotlib import pyplot as plt

lo=np.array([0, 100, 50])
hi=np.array([20, 255, 255])
color_infos=(0, 255, 255)
kernel_size=5
kernel=np.ones((5,5),np.uint8)
#cap=cv2.VideoCapture(0)
frame = cv2.imread('map_obstacles_AD.jpg')
avg_img   = cv2.blur(frame,(kernel_size,kernel_size))
plt.imshow(avg_img[:,:,::-1])
plt.figure()
#frame= cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

#gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
#ret, frame=cap.read()
image=cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
mask=cv2.inRange(image, lo, hi)

mask2=cv2.erode(mask,kernel, iterations=2)   
plt.imshow(mask2)
mask3=cv2.dilate(mask2, kernel, iterations=2 )  
image2=cv2.bitwise_and(frame, frame, mask=mask)

plt.imshow(frame[:,:,::-1])
plt.figure()
plt.imshow(image2[:,:,::-1])
plt.figure()
plt.imshow(mask3, cmap = 'gray')
#mask.shape[1]

#cv2.imshow('Camera', frame)
#cv2.imshow('image2', image2)
#cv2.imshow('Mask', mask)
```

```python
gray = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
circles = cv2.HoughCircles(gray,cv2.HOUGH_GRADIENT,1,20,param1=100,param2=10,minRadius=0,maxRadius=100)
#print(circles[0, 0])
wanted_circle = circles[0, 0]
cv2.circle(image2, (int(wanted_circle[0]), int(wanted_circle[1])), int(wanted_circle[2]), (0, 255, 0), 2)
plt.imshow(image2[:,:,::-1])
```

```python
#source : https://www.pyimagesearch.com/2016/02/08/opencv-shape-detection/#pyis-cta-modal

from shapedetector import ShapeDetector

cnts, h = cv2.findContours(mask3, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)

img_cnts = np.zeros(frame.shape)
cv2.drawContours(img_cnts, cnts, -1,(0,255,0),3)
plt.imshow(img_cnts )

#cnts = imutils.grab_contours(cnts)
sd = ShapeDetector()
```

```python
ratio =1
for c in cnts:
    # compute the center of the contour, then detect the name of the
    # shape using only the contour
    M = cv2.moments(c)
    cX = int((M["m10"] / M["m00"]) * ratio)
    cY = int((M["m01"] / M["m00"]) * ratio)
    shape = sd.detect(c)
    # multiply the contour (x, y)-coordinates by the resize ratio,
    # then draw the contours and the name of the shape on the image
    c = c.astype("float")
    c *= ratio
    c = c.astype("int")
    print(shape,cX,cY)
    img_cnts = np.zeros(frame.shape)
    cv2.drawContours(img_cnts, [c], -1, (0, 255, 0), 2)
    
    # show t©he output image
    plt.imshow(img_cnts)

```
