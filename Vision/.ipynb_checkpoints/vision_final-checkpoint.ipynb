```python
#tuto4 à rajouter source

import cv2
import numpy as np
from matplotlib import pyplot as plt

lo=np.array([110, 100, 50])
hi=np.array([130, 255, 255])
color_infos=(0, 255, 255)

kernel_size=5
kernel=np.ones((5,5),np.uint8)

frame = cv2.imread('MAP_AD.jpg')
Map_blur = cv2.blur(frame,(kernel_size,kernel_size))


Map = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
mask=cv2.inRange(Map, lo, hi)

#mask2=cv2.erode(mask,kernel, iterations=2)   
#plt.imshow(mask2)
#mask3=cv2.dilate(mask2, kernel, iterations=2 )  
Map2=cv2.bitwise_and(frame, frame, mask=mask)

plt.imshow(frame[:,:,::-1])
plt.figure()
plt.imshow(Map2[:,:,::-1])
plt.figure()
plt.imshow(mask, cmap = 'gray')
```

```python
#Detect the squares to deduce the contour of the map
#source : https://www.pyimagesearch.com/2016/02/08/opencv-shape-detection/#pyis-cta-modal

from shapedetector import ShapeDetector

cnts, h = cv2.findContours(mask, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)

img_cnts = np.zeros(frame.shape)
cv2.drawContours(img_cnts, cnts, -1,(0,255,0),3)
plt.imshow(img_cnts )

#cnts = imutils.grab_contours(cnts)
sd = ShapeDetector()
```

```python
ratio =1
for c in cnts:
    # compute the center of the contour, then detect the name of the
    # shape using only the contour
    M = cv2.moments(c)
    cX = int((M["m10"] / M["m00"]) * ratio)
    cY = int((M["m01"] / M["m00"]) * ratio)
    shape = sd.detect(c)
    # multiply the contour (x, y)-coordinates by the resize ratio,
    # then draw the contours and the name of the shape on the image
    c = c.astype("float")
    c *= ratio
    c = c.astype("int")
    print(shape,cX,cY)
    img_cnts = np.zeros(frame.shape)
    cv2.drawContours(img_cnts, [c], -1, (0, 255, 0), 2)
    
    cnt = cnts[0]
    # show the output image
    area = cv2.contourArea(cnt)
    print(area)
    plt.imshow(img_cnts)
```
